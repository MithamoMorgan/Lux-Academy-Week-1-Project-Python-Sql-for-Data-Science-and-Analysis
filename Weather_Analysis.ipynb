## 1: Find all the records where the weather was exactly clear

```python
# finding all the records where where weather was exactly clear
clear_weather = weather_data[weather_data['Weather'] == 'Clear']
# Displaying the records
clear_weather
```

## 2:Find the number of times the wind speed was exactly 4 Km/hr

```python
# finding the number of times the wind speed was exactly 4km/hr
count_speed = (weather_data['Wind Speed_km/h'] == 4).sum()
# Displaying the number of times
print(f'The number of times the wind speed was exactly 4 Km/hr was {count_speed}.')
```

```python
# Answer:The number of times wind speed was 4km/hr is 474 times
```

## 3:Check if there are NULL values present in the dataset

```python
# Checking if there are null values in the dataset
null_values = weather_data.isnull().sum()
# Displaying the results
null_values
```

```python
# Answer:We do not have any column where there are null values present in the dataset
```

## 4:Rename the column "Weather" to "Weather_Condition".

```python
# Renaming "Weather" column to "Weather_condition"
weather_data.rename(columns = {'Weather':'Weather_Condition'}, inplace = True)
# Checking if renamed
print(weather_data.columns)
```

## 5:What is the mean visibility of the dataset

```python
# Finding the mean visibility
mean_visibility = weather_data['Visibility_km'].mean()
# printing the mean visibility
print(f'The mean visibility of the dataset is:{round(mean_visibility,4)} Km')
```

## 6:Find the number of records where the wind speed is greater than 24 Km/hr and visibility is equal to 25Km

```python
# Checking how many times the 2 conditions are met using lambda functions
record_count = weather_data.apply(lambda x:x['Wind Speed_km/h'] > 24 and x['Visibility_km'] == 25, axis = 1).sum()
# displaying the number of times
record_count
print(f'The number of records is {record_count}.')
```

```python
# Confirming using a function
def check_condition(x):
    return x['Wind Speed_km/h'] > 24 and x['Visibility_km'] == 25

record_count = weather_data.apply(check_condition, axis = 1).sum()
record_count
print(f'The number of records is {record_count}.')
```

## 7:What is the mean value of each column for each weather condition?

```python
# Finding the mean of each numeric column
columns_mean = weather_data.mean(numeric_only = True)
# Displaying the result
columns_mean
```

## 8:Find all instances where the weather is clear and relative humidity is greater than 50, or visibility is above 40.

```python
# Filter the data based on the conditions
filter_data = weather_data[weather_data.apply(lambda x:x['Weather_Condition'] == 'Clear' and (x['Rel Hum_%'] > 50 or x['Visibility_km'] > 40),axis = 1)]
# Display the results
filter_data
```

```python
# Checking using boolean indexing
filtered_data = weather_data[(weather_data['Weather_Condition'] == 'Clear') & ((weather_data['Rel Hum_%'] > 50)| (weather_data['Visibility_km'] > 40))]
# Displaying the results
filtered_data
```

## 9:Find the number of weather conditions that included snow

```python
# Finding the number of weather conditions that include snow
include_snow = weather_data['Weather_Condition'].str.contains('snow', case = False).sum()
# Displaying the results
print(f'The number of weather conditions that included snow are {include_snow}.')
```
